1: INSPECCIONAR DATOS

* .head()	Muestra las primeras 5 filas, EJEMPLO=df.head(10)
* .tail()	Muestra las últimas 5 filas, EJEMPLO=	df.tail()
* .info()	Info de columnas, tipos de datos y nulos, EJEMPLO=	df.info()
* .describe()	Estadísticas descriptivas (count, mean, std, etc.), EJEMPLO=	df.describe()
* .shape	Devuelve (filas, columnas), EJEMPLO=	print(df.shape)
* .columns	Lista los nombres de las columnas, EJEMPLO=	df.columns.tolist()
* .dtypes	Muestra los tipos de datos de cada columna, EJEMPLO=	df.dtypes

```python
print("Primeras 5 filas: ")
print(muestra.head()) #impresion
print("\nUltimas 5 filas: ")
print(muestra.tail()) #como dataframe solo con muestra.tail()
print("\nEstadisticas descriptivas: ")
print(muestra.describe()) #estadisticas
print("\nInfo: ")
print(muestra.info()) #tipos de datos y nulos
print("\nShape: ")
print(muestra.shape) #filas y columnas
print("\nColumnas nombres: ")
print(muestra.columns) #nombres de las columnas
print("\nTipos de datos: ")
print(muestra.dtypes)
```

2: LIMPIAR DATOS

* .drop()	Elimina filas o columnas	df.drop('columna', axis=1)
* .dropna()	Elimina filas/columnas con valores nulos	df.dropna()
* .fillna()	Rellena valores nulos	df.fillna(0)
* .rename()	Renombra columnas	df.rename(columns={'viejo':'nuevo'})
* .astype()	Cambia el tipo de datos	df['columna'].astype(int)

```python
muestra.head()
```

```python
#METODO RENAME
muestra=muestra.rename(columns={'weather':'clima'})
muestra.head()
```

```python
#DROPNA
muestra=muestra.dropna()
muestra.head()
```

3:SELECCIONAR DATOS

* [] (indexing)	Selecciona columnas	df['columna']
* .loc[]	Selecciona por ETIQUETA (label)	df.loc[0, 'columna']
* .iloc[]	Selecciona por POSICIÓN (índice)	df.iloc[0, 1]
* .query()	Filtra con expresión string	df.query('edad > 30')

```python
print(muestra.head())
print("\nPor columna:")
print(muestra['wind'])
print("\nPor elemento(este es el elemento  de la columna wind):")
print(muestra.loc[4,'wind'])
print("\nPor indice:")
print(muestra.iloc[0,4])  #por indices de fila,columna
```

4: FILTRAR Y ORDENAR

* [df['col'] > valor]	Filtro booleano	df[df['edad'] > 30]
* .isin()	Filtra con lista de valores	df[df['ciudad'].isin(['Madrid','Barcelona'])]
* .sort_values()	Ordena por valores	df.sort_values('edad', ascending=False)
* .unique()	Valores únicos de una columna	df['ciudad'].unique()

```python
print(muestra.head())
print("\nValores unicos de clima:")
print(muestra['clima'].unique())
print("\nDe mayor a menor temperatura:")
print(muestra.sort_values('temp_max', ascending=False))
print("\nLos que tienen clima lluvioso:")
print(muestra['clima'].isin(['rain']))
```

5: AGRUPAR y AGREGAR:

* .groupby()	Agrupa por columnas	df.groupby('ciudad')
*.agg()	Aplica funciones de agregación	df.groupby('ciudad').agg({'ventas': 'sum'})
* .value_counts()	Cuenta valores únicos	df['ciudad'].value_counts()
* .pivot_table()	Crea tablas pivote	pd.pivot_table(df, values='ventas', index='ciudad')

```python
print(muestra.head())
print("\nAgrupado por clima:")
print(muestra.groupby('clima').agg({'temp_max': 'mean'}))
print("\nValores unicos de clima:")
print(muestra['clima'].value_counts())
print("\n")
print(pd.pivot_table(muestra, values='temp_max', index='wind'))
```
